```python
#We can't trust the mean value because it can get affected by outliers
#so we can use median but then also the medain of whole dataset will able to solve the problem
#we will filter each category and then fill the null values of each category with their respective median 
#with this way we will handle the null values more efficiently
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.groupby('Category')['Review Rating'].median()
```

```python
df.columns
```

```python
#Converting columns_names into lowercase and removing spaces which will helps us from error 
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
print(df.columns)
```

```python
#renaming "purchase_amount_(usd)" with "purchase_amount"
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
print(df.columns)
```

```python
#Creating a new column age_group
labels = ['Young_Adult', 'Adult', 'Middle_Age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age', 'age_group']].head(10)
```

```python
#creating a column purchase_frequency_days with the help of the column "Frequency_of_Purchase"
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Bi-Weekly' : 14,
    'Monthly' : 30,
    'Every 3 Months' : 90,
    'Quarterly' : 90,
    'Annually' : 365
}
    
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
#Now we will chwck the if any column is redundant or not-- like we have discount_applied and promo_code_used
(df['discount_applied'] == df['promo_code_used']).all()
#True both columns have same values
```

```python
#so both the columns have same values we will drop a column
df = df.drop('promo_code_used', axis = 1)
```
